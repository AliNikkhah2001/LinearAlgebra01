# Question 1 ``(30 pt.)``
* Fit a polynomial of degree $K$ to all columns of the following dataset
to predict number of "Cars" and "Buses" in the years of $2016$,$2017$,$2018$,$2019$ and $2020$.


$$
y_i = \sum_{k=0}^K w_k x_{i}^k + \epsilon_i
$$


Change the value of $K$ from 3 to 5. For each degree, plot the data and your model fit and compare the results.

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import mean_squared_error
```

```python

dataset = pd.read_csv('arac.csv', sep=';')
BaseYear = 1966

x = np.matrix(dataset.Year[0:] - BaseYear).T
cars = np.matrix(dataset.Car[0:]).T
buses = np.matrix(dataset.Bus[0:]).T

# Define degrees of polynomials
degrees = [1,2,3, 4, 5,6,7]
```

```python
dataset.info()
dataset.sample(5)
```

```python

# Define target years for prediction and adjust for base year
TargetYears = np.array([year - BaseYear for year in [2016, 2017, 2018, 2019, 2020]])

# Function to fit and predict based on polynomial degree
def fit_and_predict(x, y, target_years, degree):
    # Create Vandermonde matrix for the given data
    A = np.column_stack([np.power(x, i) for i in range(degree + 1)])
    
    # Find the weight matrix using least squares
    W, _, _, _ = np.linalg.lstsq(A, y, rcond=None)
    
    # Compute the fitted values for the original data
    f = A @ W
    
    # Create Vandermonde matrix for the target years
    A_target = np.column_stack([np.power(target_years, i) for i in range(degree + 1)])
    
    # Predict values for the target years
    f_target = A_target @ W
    
    return f, f_target
degrees = range(1, 11)
# Loop through different polynomial degrees
for K in range(1, 11):
    # Fit and predict for cars
    f_cars, f2_cars = fit_and_predict(x, cars, TargetYears, K)

    # Plot results for cars
    plt.figure(figsize=(10, 5))
    plt.plot(x + BaseYear, cars, 'o-', label='Actual Cars')
    plt.plot(x + BaseYear, f_cars, 'r', label='Fitted Cars')
    plt.plot(TargetYears + BaseYear, f2_cars, 'ro-', label=f'Predicted Cars (Degree {K})')
    plt.legend()
    plt.title(f'Polynomial Fitting for Cars (Degree {K})')
    plt.xlabel('Year')
    plt.ylabel('Number of Cars')
    plt.grid(True)
    plt.show()

    print(f"Predicted Numbers for Cars (Degree {K}):")
    print(f2_cars)

    # Fit and predict for buses
    f_buses, f2_buses = fit_and_predict(x, buses, TargetYears, K)

    # Plot results for buses
    plt.figure(figsize=(10, 5))
    plt.plot(x + BaseYear, buses, 'o-', label='Actual Buses')
    plt.plot(x + BaseYear, f_buses, 'r', label='Fitted Buses')
    plt.plot(TargetYears + BaseYear, f2_buses, 'ro-', label=f'Predicted Buses (Degree {K})')
    plt.legend()
    plt.title(f'Polynomial Fitting for Buses (Degree {K})')
    plt.xlabel('Year')
    plt.ylabel('Number of Buses')
    plt.grid(True)
    plt.show()

    print(f"Predicted Numbers for Buses (Degree {K}):")
    print(f2_buses)
```

As you learned in the course, the least squares method is a way of estimating a function given sample points $$(x_i, y_i), \quad  0 \leq i \leq n$$ with the assumption of linearity of the underlying function.
### step 1
Create n=10 sample points $(x_i, y_i)$ where $$y_i = 2x_i + 3 + \epsilon_i$$ $\epsilon_i$ is an error term (measurement error etc.) added to $y_i$ and is modeled by random variables: $\epsilon_i \stackrel{iid}{\sim} \mathcal{N}(0, 1) $  . ### step 2
Given the samples of the previous part, try to estimate a linear function that models the relationship between $x_i$'s and $y_i$'s using the least squares method (find a and b such that the error function $\sum_{i=0}^{n-1} ||ax_i + b - y_i||^2$ is minimized)

Plot the sample points and the line you estimated in the previous part in the same graph using matplotlib.
### step 4
Plot the error function $\sum_{i=0}^{n-1} ||ax_i + b - y_i||^2$ as a function of n where $10\leq n \leq 50$ and a and b are the least squares estimates of the line $ax + b = y$.
### step 5
Repeat part 1 but this time create n=20 sample points with a quadratic term added to y: $$y_i = x^2_i + 2x_i + 3 + \epsilon_i$$ $\epsilon_i$ has the same distribution as before and x's range from 0 to 5 likewise.
### step 6
Try to use the least squares method to estimate the coefficients of the quadratic function given the sample points of part 5 (hint: in the least squares method we minimze the norm of the error vector $||Ax-y||^2$. ### step 7
Repeat part 3 for the estimated quadratic function of the previous part.

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create sample points for a linear function with noise
def generate_linear_points(n):
    np.random.seed(0)
    x = np.linspace(0, 5, n)
    noise = np.random.normal(0, 1, n)
    y = 2 * x + 3 + noise
    return x, y

# Step 2: Estimate linear coefficients using least squares method
def estimate_linear(x, y):
    A = np.vstack([x, np.ones(len(x))]).T
    a, b = np.linalg.lstsq(A, y, rcond=None)[0]
    return a, b

# Step 3: Plot the linear samples and estimated line
def plot_linear(x, y, a, b):
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, 'o', label='Sample Points')
    plt.plot(x, a * x + b, 'r', label=f'Estimated Line: y = {a:.2f}x + {b:.2f}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Linear Estimation')
    plt.grid(True)
    plt.legend()
    plt.show()

# Step 4: Plot error function as a function of n
def plot_error_vs_n():
    errors = []
    ns = range(10, 51)
    for n in ns:
        x, y = generate_linear_points(n)
        a, b = estimate_linear(x, y)
        predicted = a * x + b
        error = np.sum((predicted - y) ** 2)
        errors.append(error)

    plt.figure(figsize=(8, 6))
    plt.plot(ns, errors, 'o-', label='Error Function')
    plt.xlabel('n')
    plt.ylabel('Sum of Squared Errors')
    plt.title('Error Function vs Sample Size (n)')
    plt.grid(True)
    plt.legend()
    plt.show()

# Step 5: Create sample points for a quadratic function with noise
def generate_quadratic_points(n):
    np.random.seed(0)
    x = np.linspace(0, 5, n)
    noise = np.random.normal(0, 1, n)
    y = x ** 2 + 2 * x + 3 + noise
    return x, y

# Step 6: Estimate quadratic coefficients using least squares method
def estimate_quadratic(x, y):
    A = np.vstack([x ** 2, x, np.ones(len(x))]).T
    coef = np.linalg.lstsq(A, y, rcond=None)[0]
    return coef

# Step 7: Plot the quadratic samples and estimated curve
def plot_quadratic(x, y, coef):
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, 'o', label='Sample Points')
    predicted = coef[0] * x ** 2 + coef[1] * x + coef[2]
    plt.plot(x, predicted, 'r', label=f'Estimated Curve: y = {coef[0]:.2f}x + {coef[1]:.2f}x + {coef[2]:.2f}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Quadratic Estimation')
    plt.grid(True)
    plt.legend()
    plt.show()

```

A structure's various components can be represented by a single **state vector**, which changes over time due to certain processes. This changing state vector, when modeled using linear transformations at discrete intervals, forms the basis of a **discrete dynamical system**.

Starting with an initial state $X_0$, transitions to subsequent states are governed by a matrix $P$, embodying linear transformations. For instance, $X_1 = PX_0$ marks the first transition, leading to a sequence where $X_2 = P^2X_0$, $X_3 = P^3X_0$, and so forth, generalizing to $X_t = PX_{t-1}$ for each time step $t$.

This approach is particularly effective in modeling web navigation, where each webpage's visit probability at any given time is captured by the state vector $X_t$. The transition matrix $P$ specifies the likelihood of moving from one page to another, with each element $p_{ij}$ representing the probability of transitioning from page $j$ to page $i$. To ensure the model's consistency, the sum of probabilities in each column of $P$ must equal one, reflecting the comprehensive distribution of navigating from any given page to the next.

To illustrate with a small set of pages, say $n=5$, we construct matrix $P$ adhering to the principle that all diagonal elements are $0$, emphasizing no page links back to itself immediately. Additionally, to ensure that navigation probabilities from one page to another are properly distributed, the sum of the entries in each column of $P$ must equal one. Here's how such a matrix might look for $n=5$:

$$P=
\begin{bmatrix}
0 & 0.2 & 0.3 & 0.4 & 0.1 \\
0.3 & 0 & 0.2 & 0.1 & 0.4 \\
0.2 & 0.3 & 0 & 0.2 & 0.3 \\
0.1 & 0.2 & 0.4 & 0 & 0.1 \\
0.4 & 0.3 & 0.1 & 0.3 & 0.1
\end{bmatrix}$$

This configuration ensures a balanced probability distribution for transitions between pages, making our web navigation model realistic.

For the initial state vector $X_0$, where the browser starts at page $4$, $X_0$ would be a vector with its fourth entry as one and all others as zero, representing a 100% probability of starting from page 4.

Write a function to compute the navigation probabilities to all five pages after t transition. Then use the function to calculate a new vector $X_t$, representing the probabilities that the browser is on each of the five pages after following t links from the starting page. At last apply the function on $X_0$ assuming symmetric transition probabilities and t to be 10.

```python
import numpy as np

P = np.array([
    [0, 0.2, 0.3, 0.4, 0.1],
    [0.3, 0, 0.2, 0.1, 0.4],
    [0.2, 0.3, 0, 0.2, 0.3],
    [0.1, 0.2, 0.4, 0, 0.1],
    [0.4, 0.3, 0.1, 0.3, 0.1]
])

X_0 = np.array([0, 0, 0, 1, 0])  # Starting from page 4
```

```python
def compute_navigation_probabilities(P, X_0, t):
    X_t = np.linalg.matrix_power(P, t) @ X_0
    return X_t

t = 10
X_t = compute_navigation_probabilities(P, X_0, t)
print("Probabilities after", t, "transitions:", X_t)
```

Predicting long-term visit probabilities, represented by $X_m$ for a large $m$, reveals which pages are likely to be frequented after extensive random navigation. However, repeatedly applying transformations to achieve this is computationally intensive. Leveraging eigenvalues and eigenvectors can substantially streamline the process, significantly minimizing the computational workload.

Find eigenvalues and eigenvectors for the previously defined matrix $P$.

```python

# Calculate eigenvalues and eigenvectors of P
eigenvalues, eigenvectors = np.linalg.eig(P)

# Display the eigenvalues
print("Eigenvalues of P:")
print(eigenvalues)

# Display the eigenvectors
print("\nEigenvectors of P:")
print(eigenvectors)

# Find the eigenvector associated with the eigenvalue 1
# It may require normalization and adjustment to ensure all entries are non-negative
for i, val in enumerate(eigenvalues):
    if np.isclose(val, 1):
        # Normalize the eigenvector to make it a valid probability vector (sum equals 1)
        steady_state_vector = np.abs(eigenvectors[:, i]) / np.sum(np.abs(eigenvectors[:, i]))
        print("\nSteady state vector (long-term visit probabilities):")
        print(steady_state_vector)
        break
```

A **Markov matrix** is characterized by its square shape, nonnegative entries, and the property that the sum of the elements in each column equals $1$. A key attribute of Markov matrices is that they always possess an eigenvalue of $1$. 

Given that matrix $P$ is designed as a Markov matrix, it naturally includes an eigenvalue of $1$. This feature allows us to express the equation $X_m = PX_{m-1}$ in a simplified form as $PX_{m-1} = 1 \times X_m$. For long-term probability predictions within $X_m$, identifying an eigenvector associated with the eigenvalue $1$ is crucial. This is because it leads to the simplified equation $PX = X$, facilitating the determination of steady-state probabilities without extensive computations.

After doing so check if your answer satisfies the equation with proper code. (use np.isclose)

```python
for i in range(len(eigenvalues)):
    print(np.isclose(P @ eigenvectors[:, i], eigenvalues[i] * eigenvectors[:, i]))
```

```python

# Calculate eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(P)

# Identify the eigenvector associated with the eigenvalue 1
index = np.where(np.isclose(eigenvalues, 1))[0][0]  # Find index where eigenvalue is close to 1
steady_state_vector = np.real(eigenvectors[:, index])  # Extract the corresponding eigenvector
steady_state_vector = steady_state_vector / np.sum(steady_state_vector)  # Normalize

# Check if PX = X
PX = P @ steady_state_vector
is_steady_state = np.isclose(PX, steady_state_vector)

print("Steady State Vector:")
print(steady_state_vector)
print("\nPX:")
print(PX)
print("\nIs PX equal to X?")
print(is_steady_state)

# Overall check if all entries are close to each other
if np.all(is_steady_state):
    print("\nThe computed steady state vector satisfies PX = X.")
else:
    print("\nThe computed steady state vector does not satisfy PX = X.")
```


The transition matrix \(P\) is defined as follows for \(n=5\) pages:
\begin{aligned}
P = &\begin{bmatrix}
0 & 0.2 & 0.3 & 0.4 & 0.1 \\

0.2 & 0.3 & 0 & 0.2 & 0.3 \\
0.1 & 0.2 & 0.4 & 0 & 0.1 \\
0.4 & 0.3 & 0.1 & 0.3 & 0.1

In a more realistic model of the web, some pages may not have outbound links (dangling nodes), and users may sometimes choose to navigate to a random page rather than following an outbound link. The Google PageRank algorithm introduces a damping factor to model this behavior, making the Markov process more reflective of real-world web navigation.

Dangling nodes are pages without outbound links. These nodes would lead to columns of zeros in the transition matrix, which is problematic for calculating the steady state. To address this, distribute the probability mass of dangling nodes equally among all pages.

First identify dangling nodes then djust the transition matrix $P$ by redistributing the dangling nodes' probability mass equally across all pages, ensuring every column sums to 1.

```python

# Define the transition matrix P, including a dangling node
P = np.array([
    [0, 0, 0, 0, 0],     # Probabilities from each page to Page 1
    [1, 0, 0.5, 0, 0],   # Probabilities from each page to Page 2
    [0, 0, 0, 0.5, 0],   # Probabilities from each page to Page 3
    [0, 1, 0.5, 0, 0],   # Probabilities from each page to Page 4
    [0, 0, 0, 0.5, 0]    # Page 5 is a dangling node
])

# Number of pages
n = P.shape[1]

# Identify dangling nodes (columns that sum to 0)
dangling_nodes = np.where(np.sum(P, axis=0) == 0)[0]

# Redistribute the dangling nodes' probability mass equally across all pages
for j in dangling_nodes:
    P[:, j] = 1 / n  # Distribute 1/n to each page

# Print the adjusted transition matrix P
print("Adjusted Transition Matrix P:")
print(P)

# Validate that each column now sums to 1
column_sums = np.sum(P, axis=0)
print("\nColumn sums (should be all 1s):")
print(column_sums)
```

The damping factor models the behavior of web surfers who occasionally jump to a random page instead of following outbound links. This factor, usually set to 0.85, balances between following links ($P$) and jumping at random ($\frac{1}{n}$, where \(n\) is the number of pages).

Apply the damping factor $d$ to the transition matrix $P$ to get $P' = dP + \frac{1-d}{n}E$, where $E$ is an $n \times n$ matrix with all elements equal to $\frac{1}{n}$.

```python
import scipy.sparse.linalg as spla
# Example transition matrix P with 5 pages
P = np.array([
    [0, 0, 0, 0, 0.2],  # Probabilities from each page to Page 1
    [1, 0, 0.5, 0, 0.2],  # Probabilities from each page to Page 2
    [0, 0, 0, 0.5, 0.2],  # Probabilities from each page to Page 3
    [0, 1, 0.5, 0, 0.2],  # Probabilities from each page to Page 4
    [0, 0, 0, 0.5, 0.2]   # Probabilities from each page to Page 5 (adjusted for dangling node)
])

n = P.shape[0]  # Number of pages
d = 0.85  # Damping factor

# Create matrix E where every element is 1/n
E = np.ones((n, n)) / n

# Apply damping factor to transition matrix P
P_prime = d * P + (1 - d) * E

# Print the adjusted transition matrix P'
print("Adjusted Transition Matrix P' with Damping Factor:")
print(P_prime)
```

The steady state distribution with the damping factor reflects a more realistic scenario of web page navigation. Finding this distribution involves solving for the principal Eigenvector of the modified transition matrix $P'$.

```python
# Compute eigenvalues and eigenvectors of P'
eigenvalues, eigenvectors = np.linalg.eig(P_prime.T)  # .T because we deal with right eigenvectors

# Find the index of the eigenvalue that is closest to 1
index = np.argmin(np.abs(eigenvalues - 1))

# Extract the corresponding eigenvector and normalize it
principal_eigenvector = np.real(eigenvectors[:, index])
normalized_eigenvector = principal_eigenvector / np.sum(principal_eigenvector)

# Display the normalized principal eigenvector
print("Normalized Principal Eigenvector (Steady-State Distribution):")
print(normalized_eigenvector)
```

Real-world web link structures can be vast, resulting in large, sparse transition matrices. Direct computation methods become impractical, necessitating efficient computational strategies.

Represent the transition matrix $P$ and modifications using sparse matrix techniques to handle large datasets. Then implement an iterative method, such as the power iteration algorithm, to find the principal Eigenvector efficiently.

```python
import numpy as np
from scipy.sparse import csr_matrix
from scipy.sparse.linalg import eigs

def create_sparse_transition_matrix(P_dense):
    """Convert dense matrix P to a sparse CSR format."""
    return csr_matrix(P_dense)

def power_iteration(P, num_iterations=100, d=0.85):
    """Compute the principal eigenvector using the power iteration method."""
    n = P.shape[0]
    E = np.ones((n, n)) / n
    P_damped = d * P + (1 - d) * csr_matrix(E)
    
    # Start with an initial vector with equal probabilities
    x = np.ones(n) / n
    
    # Power iteration
    for _ in range(num_iterations):
        x_new = P_damped @ x
        x_new /= np.linalg.norm(x_new, 1)  # Normalize the vector to ensure it sums to 1
        if np.allclose(x_new, x):
            break
        x = x_new
    
    return x

# Example dense transition matrix P
P_dense = np.array([
    [0.0, 0.2, 0.3, 0.4, 0.1],
    [0.3, 0.0, 0.2, 0.1, 0.4],
    [0.2, 0.3, 0.0, 0.2, 0.3],
    [0.1, 0.2, 0.4, 0.0, 0.1],
    [0.4, 0.3, 0.1, 0.3, 0.1]
])

# Convert to sparse matrix
P_sparse = create_sparse_transition_matrix(P_dense)

# Apply power iteration to find the principal eigenvector
principal_eigenvector = power_iteration(P_sparse)

print("Principal Eigenvector (Steady-State Distribution):")
print(principal_eigenvector)
```

Auto-regressive (AR) models are fundamental in signal processing for modeling time series data. An AR model predicts future values based on a linear combination of past values. Mathematically, it's represented as 
$$
\hat{y}[t] = \sum_{i=1}^{p} a_i y[t-i]
$$, 
where $\hat{y}[t]$ is the predicted value at time $t$, $y[t-i]$ are the past values, and $a_i$ are the coefficients that need to be determined. Cross-regressive models extend this concept by modeling the relationship between two different time series, predicting values of one series using past values of another.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load data
qbo_data = np.loadtxt('qbo.txt')
series_1 = qbo_data[:, 0]  # Assuming the first column represents the first time series

# Set AR model order `p`
p = 5

# Prepare AR data
def prepare_ar_data(series, p):
    """ Prepares the input and target arrays for AR modeling """
    n = len(series)
    X = np.zeros((n - p, p))
    y = np.zeros(n - p)
    for i in range(p, n):
        X[i - p] = series[i - p:i]
        y[i - p] = series[i]
    return X, y

# Prepare the data
X, y = prepare_ar_data(series_1, p)

# Fit the AR model using linear regression
model = LinearRegression()
model.fit(X, y)
coefficients = model.coef_
intercept = model.intercept_

# Predict the values using the trained AR model
y_pred = model.predict(X)

# Plot the original and predicted series
plt.figure(figsize=(10, 6))
plt.plot(np.arange(len(series_1)), series_1, label='Original Series')
plt.plot(np.arange(p, len(series_1)), y_pred, label=f'AR Model (p={p})', linestyle='dashed')
plt.xlabel('Time Index')
plt.ylabel('Value')
plt.title('Auto-Regressive Model Prediction')
plt.legend()
plt.grid(True)
plt.show()

# Print AR model coefficients
print(f"AR Model Coefficients (p={p}): {coefficients}")
print(f"Intercept: {intercept}")
```

Construct an embedding matrix from lagged versions of the data series. This matrix, often denoted as $M$, captures the structure of dependencies in past values. Use the pseudo-inverse of $M$ to solve for the coefficients $a_i$ that minimize the prediction error in a least squares sense.

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data
qbo_data = np.loadtxt('qbo.txt')
series_1 = qbo_data[:, 0]  # Assuming the first column represents the first time series

# Define the order of the model `p`
p = 5

# Function to build the embedding matrix `M`
def build_embedding_matrix(series, p):
    """ Builds an embedding matrix `M` from lagged versions of the time series `series` """
    n = len(series)
    M = np.zeros((n - p, p))
    for i in range(p, n):
        M[i - p] = series[i - p:i]
    return M

# Build the embedding matrix `M`
M = build_embedding_matrix(series_1, p)

# Prepare the target vector `y`
y = series_1[p:]

# Calculate the pseudo-inverse of `M`
M_pseudo_inv = np.linalg.pinv(M)

# Solve for the coefficients `a_i` using the pseudo-inverse
coefficients = M_pseudo_inv @ y

# Predict the future values using the coefficients
y_pred = M @ coefficients

# Plot the original and predicted series
plt.figure(figsize=(10, 6))
plt.plot(np.arange(len(series_1)), series_1, label='Original Series')
plt.plot(np.arange(p, len(series_1)), y_pred, 'r--', label=f'Predicted Series (p={p})')
plt.xlabel('Time Index')
plt.ylabel('Value')
plt.title('Auto-Regressive Model Prediction Using Pseudo-Inverse')
plt.grid(True)
plt.legend()
plt.show()

# Print coefficients of the AR model
print(f"AR Model Coefficients (pseudo-inverse): {coefficients}")
```

In the cross-regression model described by the equation

$$
\hat{y}[t] = \sum_{i=1}^{p} a_i z[t-i]
$$

the goal is to predict the current value of the time series \(y\) based on the past values of another time series \(z\). Here, \(p\) represents the number of past values (lags) of \(z\) used in the prediction, and \(a_i\) are the coefficients that quantify the impact of these past values on \(y\).

### Analyzing the Coefficients

The coefficients \(a_i\) provide a quantitative measure of how strongly each past value of \(z\) influences the current value of \(y\). The key points to consider include:

- **Magnitude of Coefficients:** Larger absolute values of the coefficients \(a_i\) indicate a stronger influence of the corresponding lagged value of \(z\) on \(y\). If a coefficient is significantly larger than others, it suggests that the corresponding lagged time point of \(z\) has a more substantial impact on predicting \(y\).

- **Sign of Coefficients:** The sign of each coefficient indicates the nature of the relationship:
  - A positive coefficient (\(+\)) suggests that as \(z[t-i]\) increases, \(y[t]\) also increases, implying a direct relationship.
  - A negative coefficient (\(-\)) suggests an inverse relationship, where increasing values of \(z[t-i]\) lead to decreases in \(y[t]\).

### Implications for Interaction and Influence

The analysis of these coefficients can reveal not only the strength but also the potential direction of influence between the two time series:

1. **Strong Interactions:** If several coefficients are significantly different from zero, it indicates strong interactions between \(z\) and \(y\). This could suggest that past trends or patterns in \(z\) reliably influence future values of \(y\).

2. **Direction of Influence:** While cross-regression by itself does not confirm causality, significant coefficients can indicate a directional influence where changes in \(z\) precede and predict changes in \(y\). This is particularly relevant in fields like economics or environmental science, where predictive relationships can inform policy or operational decisions.

3. **Analyzing Multiple Time Lags:** By examining which lags have the most substantial coefficients, you can determine specific times when \(z\) most impacts \(y\). For instance, a large coefficient at a lag of three months might suggest seasonal effects from \(z\) on \(y\).

Cross-regression provides valuable insights into how variables interact over time. However, it's essential to remember that significant correlations revealed by cross-regression do not imply causation. Cross-regression is a statistical approach that models the relationship between two time series, where the objective is to predict the current value of one series based on past values of another. This method can reveal the extent and nature of influence that one time series may have on another. In mathematical terms, this is often structured as follows:

### Mathematical Representation

Given two time series \(y\) and \(z\), a cross-regression model can be formulated as:

$$
\hat{y}[t] = \sum_{i=1}^{p} a_i z[t-i]
$$

Here:
- \( \hat{y}[t] \) is the predicted value of the time series \(y\) at time \(t\).
- \( z[t-i] \) represents the values of another time series \(z\) at past times \(t-i\), where \(i\) ranges from 1 to \(p\).
- \( a_i \) are the coefficients to be estimated, showing how previous values of \(z\) influence the current value of \(y\).
- \( p \) is the number of past points (lags) of \(z\) used to predict \(y\).

### Interpretation and Analysis

The coefficients \(a_i\) provide insights into the dynamics between the two series:
- **Magnitude of Coefficients:** Larger absolute values of \(a_i\) indicate a stronger influence of \(z[t-i]\) on \(y[t]\). Positive values suggest a direct relationship, while negative values suggest an inverse relationship.
- **Sign of Coefficients:** The sign of each coefficient tells whether the effect is to increase (\(+\)) or decrease (\(-\)) \(y[t]\) as \(z[t-i]\) increases.

### Practical Steps for Analysis

1. **Data Preparation:** Organize your data so that each observation of \(y\) corresponds to several lagged observations of \(z\). This is often done by constructing an embedding matrix where each row contains the lagged values of \(z\) for a particular time \(t\).

2. **Model Estimation:** Use statistical or machine learning methods to estimate the coefficients \(a_i\). Methods like ordinary least squares, ridge regression, or even machine learning algorithms can be applied depending on the data characteristics.

3. **Evaluation of Fit:** Assess the model using criteria like the Mean Squared Error (MSE) or the coefficient of determination (\(R^2\)). These metrics evaluate how well the model captures the relationship between \(y\) and \(z\).

4. **Hypothesis Testing:** If the analysis is being done in a scientific or econometric context, statistical tests (like t-tests on coefficients) can determine the significance of the relationships.

### Example Applications

- **Economics:** Predicting consumer spending based on economic indices.
- **Environmental Science:** Modeling pollution levels based on traffic data.
- **Finance:** Forecasting stock prices based on economic indicators.

### Conclusion

Cross-regression can be a powerful tool in time series analysis, allowing analysts to uncover and quantify the relationships between different series. By understanding these dynamics, more accurate predictions and better strategic decisions can be made.

Formulate the inverse problem as a least squares problem. Then solve the inverse problem to get the reconstructed image. Finally display the original, filtered and the reconstructed images in a subplot. Use $\textbf{lsqr}$ from $\textbf{scipy.sparse.linalg}$ to solve least squares knowing that L is in sparse format. 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.sparse.linalg import lsqr
from scipy.sparse import csr_matrix
import networkx as nx

# Create the Laplacian matrix using networkx for a grid graph
def create_laplacian_matrix(n):
    G = nx.grid_2d_graph(n, n)
    return nx.laplacian_matrix(G)

# Image dimensions
n = 128

# Assume resized_image is loaded as a 128x128 numpy array
# Flatten the image
x = resized_image.flatten()

# Generate the Laplacian matrix
L = create_laplacian_matrix(n)

# Apply the Laplacian filter
x_filtered = L.dot(x)

# Solve the inverse problem using least squares
# Convert L to a CSR (Compressed Sparse Row) format for efficient calculations
L_csr = csr_matrix(L)
result = lsqr(L_csr, x_filtered)
x_reconstructed = result[0]

# Reshape the outputs back to 2D for visualization
x_filtered_image = x_filtered.reshape((n, n))
x_reconstructed_image = x_reconstructed.reshape((n, n))

# Plot the original, filtered, and reconstructed images
plt.figure(figsize=(18, 6))
plt.subplot(1, 3, 1)
plt.imshow(resized_image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(x_filtered_image, cmap='gray')
plt.title('Filtered Image')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(x_reconstructed_image, cmap='gray')
plt.title('Reconstructed Image')
plt.axis('off')

plt.show()
```
